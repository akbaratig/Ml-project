# Histogram

```python
fig, axs = plt.subplots(nrows=2, ncols=7, figsize=(20, 10))
axs = axs.flatten()
index = 0

for k, v in df.items():
    axs[index].hist(v, bins=15)
    axs[index].set_title(k)
    index += 1
    
plt.tight_layout(pad=0.4, w_pad=0.4, h_pad=5.0)
```
