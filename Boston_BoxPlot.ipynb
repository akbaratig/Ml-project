# BoXPlot

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axs = plt.subplots(ncols=7, nrows=2, figsize=(20, 10))
axs = axs.flatten()
index = 0

for k, v in df.items():
    sns.boxplot(y=k , data=df, ax=axs[index])
    index += 1
plt.tight_layout(pad=0.4, w_pad=0.4, h_pad=5.0)
```
