# BoXPlot

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axs = plt.subplots(ncols=7, nrows=2, figsize=(20, 10))
axs = axs.flatten()
index = 0

for k, v in df.items():
    sns.boxplot(y=k , data=df, ax=axs[index])
    index += 1
plt.tight_layout(pad=0.4, w_pad=0.4, h_pad=5.0)
```

```python
df = df[~(df['medv'] >= 50)]
print(df.shape)
```

# Histogram

```python
fig, axs = plt.subplots(nrows=2, ncols=7, figsize=(20, 10))
axs = axs.flatten()
index = 0

for k, v in df.items():
    axs[index].hist(v, bins=15)
    axs[index].set_title(k)
    index += 1
    
plt.tight_layout(pad=0.4, w_pad=0.4, h_pad=5.0)
```

# Distribution

```python
fig, axs = plt.subplots(nrows=2, ncols=7, figsize=(20, 10))
index = 0
axs = axs.flatten()

for k, v in df.items():
    sns.distplot(v,  ax=axs[index], bins=15)
    index +=1
```
