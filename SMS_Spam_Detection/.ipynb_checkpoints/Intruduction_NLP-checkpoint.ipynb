# Stop Words Persian

```python
with open('stopwords.txt', encoding='utf-8') as f:
    stop = f.readlines()

stop[:10]
```

```python
stop_words_persian = [line.replace('\n', '') for line in stop]
stop_words_persian[:10]
```

```python
stop_words_persian[10:20]
```

```python
print(len(stop_words_persian))
```

# Stop Words English

```python
with open('stopwords_english.txt') as f:
    stop_en = f.readlines()

stop_en[:10]
```

```python
stop_words_engilsh = list(set([line.replace('\n', '') for line in stop_en]))
stop_words_engilsh[:10]
```

```python
len(stop_words_engilsh)
```

```python
stop_words = stop_words_persian + stop_words_engilsh
stop_words[1415:1425]
```

```python
len(stop_words)
```

# Stemming Porter

```python
from nltk.stem.porter import *
stemmer = PorterStemmer()
samples = ["connection", "connecting", "connected"]

for token in samples:
    print(token + " -> " + stemmer.stem(token))
```
